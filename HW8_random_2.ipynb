```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as sts
```

```python
norm_rv = sts.norm(loc = 0, scale = 0.5)
norm_rv1 = sts.norm(loc = 2, scale = 0.5)
norm_rv2 = sts.norm(loc = 2, scale = 1)
norm_rv3 = sts.norm(loc = 2, scale = 2)
norm_rv4 = sts.norm(loc = 2, scale = 0.3)
x = np.linspace(-4,4,100)
plt.plot(x, norm_rv.pdf(x))
plt.plot(x, norm_rv1.pdf(x))
plt.plot(x, norm_rv2.pdf(x))
plt.plot(x, norm_rv3.pdf(x))
plt.plot(x, norm_rv4.pdf(x))
plt.show()
```

```python
exp_rv = sts.expon(-2)
exp_rv1 = sts.expon(-1)
exp_rv2 = sts.expon(0)
exp_rv3 = sts.expon(1)
exp_rv4 = sts.expon(2)
x = np.linspace(0,4)
plt.plot(x, exp_rv.pdf(x))
plt.plot(x, exp_rv1.pdf(x))
plt.plot(x, exp_rv2.pdf(x))
plt.plot(x, exp_rv3.pdf(x))
plt.plot(x, exp_rv4.pdf(x))
plt.show()
```

```python
stu_rv = sts.t(0.05)
stu_rv1 = sts.t(0.1)
stu_rv2 = sts.t(0.5)
stu_rv3 = sts.t(4)
stu_rv4 = sts.t(1000)
x = np.linspace(-4,4)
plt.plot(x, stu_rv.pdf(x))
plt.plot(x, stu_rv1.pdf(x))
plt.plot(x, stu_rv2.pdf(x))
plt.plot(x, stu_rv3.pdf(x))
plt.plot(x, stu_rv4.pdf(x))
plt.show()
```
